1. What is hypothesis testing in statistics ?

Ans. Hypothesis testing is a statistical method used to make a decision or draw a conclusion about a population parameter based on sample data.It involves formulating two competing statements—the null hypothesis (H₀), which represents no effect or no difference, and the alternative hypothesis (H₁ or Hₐ), which represents the presence of an effect or difference—and using sample evidence to determine whether there is enough statistical support to reject the null hypothesis at a chosen level of significance.

2.  What is the null hypothesis, and how does it differ from the alternative
hypothesis ?

Ans. The null hypothesis is a statement that there is no effect, no difference, or no relationship between variables in a population. It serves as the default assumption that any observed difference in sample data is due to random chance.The alternative hypothesis is a statement that there is an effect, a difference, or a relationship between variables. It contradicts the null hypothesis and represents what the researcher aims to support.The null hypothesis assumes no change or no effect, while the alternative hypothesis proposes that a change or effect exists.

3.  Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Ans. The significance level is the predetermined probability of rejecting the null hypothesis when it is actually true (committing a Type I error). It is denoted by the symbol alpha (α), and common values are 0.05, 0.01, or 0.10.The significance level serves as the threshold for deciding whether the observed sample result is statistically significant. If the p-value is less than or equal to the chosen significance level (α), the null hypothesis is rejected; if the p-value is greater than α, the null hypothesis is not rejected.

4. What are Type I and Type II errors? Give examples of each.

Ans. A Type I error occurs when the null hypothesis is rejected even though it is actually true. In other words, it is a false positive result.A medical test concludes that a patient has a disease (rejects H₀), but in reality, the patient is healthy (H₀ is true).A Type II error occurs when the null hypothesis is not rejected even though it is actually false. In other words, it is a false negative result.A medical test concludes that a patient does not have a disease (fails to reject H₀), but in reality, the patient actually has the disease (H₀ is false).

5.  What is the difference between a Z-test and a T-test? Explain when to use
each.

Ans. A Z-test is a statistical test used to determine whether there is a significant difference between a sample mean and a population mean when the population standard deviation is known and the sample size is large (generally
 n≥30). It is based on the standard normal distribution (Z-distribution).A Z-test is used when the population standard deviation is known and the sample size is large, or when the sample size is large enough for the Central Limit Theorem to apply.A T-test is a statistical test used to determine whether there is a significant difference between a sample mean and a population mean when the population standard deviation is unknown and must be estimated from the sample. It is based on the t-distribution.A T-test is used when the population standard deviation is unknown and the sample size is small (generally n<30),or when working with small samples where the data are approximately normally distributed.The main difference is that a Z-test uses the known population standard deviation and the normal distribution, while a T-test uses the sample standard deviation and the t-distribution, especially for small sample sizes.

```python
''' 7.  Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9] '''

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

data = np.array(sample_data)

mu_0 = 50

n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

standard_error = sample_std / np.sqrt(n)

z_stat = (sample_mean - mu_0) / standard_error

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-Statistic:", z_stat)
print("P-Value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")

''' Interpretation of Results :- A sample mean approximately close to 50.A Z-statistic around 1 to 2 (approximately).
    A p-value greater than 0.05.


```

```python
'''  8.  Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib. '''

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
mu = 100
sigma = 15
n = 100

data = np.random.normal(mu, sigma, n)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

margin_of_error = t_critical * (sample_std / np.sqrt(n))

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label='CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label='CI Upper')

plt.title("Normal Distribution Simulation with 95% Confidence Interval")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

```python
''' 9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean. '''

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot histogram of standardized data.
    """
    data = np.array(data)

    mean = np.mean(data)
    std = np.std(data, ddof=1)

    z_scores = (data - mean) / std

    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.title("Histogram of Z-Scores (Standardized Data)")
    plt.xlabel("Z-Score")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()

    return z_scores

data = [45, 50, 55, 60, 52, 48, 49, 53, 47, 51]
z_scores = calculate_z_scores(data)

print("Z-Scores:")
print(z_scores)

''' A Z-score is a standardized value that represents the number of standard
    deviations a data point is above or below the mean of a dataset.
    Z = 0: The data point is exactly at the mean.
    Z > 0: The data point is above the mean by Z standard deviations.
    Z < 0: The data point is below the mean by |Z| standard deviations.'''
```
